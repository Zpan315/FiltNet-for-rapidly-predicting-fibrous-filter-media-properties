```python
import h5py
import matplotlib.pyplot as plt

base_dir = 'FiltNet-for-rapidly-predicting-fibrous-filter-media-properties' #Change this to the base directory of the project
Img_path = f'{base_dir}/HDF/FiltNet_1000d.h5'

Index = 12

with h5py.File(Img_path,'r') as f:
    image_array = f['X'][Index,...] # images
    X2 = f['X2'][Index,...] # metadata
    Y = f['Y'][Index,...] # Filter properties

fig, axs = plt.subplots(1, 3, figsize=(15, 8))  # Create a 3x3 grid of subplots
fig.suptitle(f"Trained Images {Index}", fontsize=16)

#for i in range(3):
for i in range(3):
    ax_original = axs[i]  
    ax_original.imshow(image_array[:, :, i], cmap='gray', vmin=0, vmax=1)
    ax_original.set_title(f"Img {i+1}")
    ax_original.axis('off') 
plt.tight_layout()
plt.show()
```
